```python
for station in ["KDFX"]: #, "KEWX", "KGRK", "KSJT"]:
    for year in [2012]: # range(2000, 2021):
        system_pred = pd.read_csv(f"../../data/texas_bats_v3_bug/tracks_{station}_{year}0101_{year}1231.txt")

        current_local_time = system_pred.local_time[0]
        current_track_id = system_pred.track_id[0]
        current_UTC_date = system_pred.filename[0][4:12]

        for i in range(1, system_pred.shape[0]):
            # same track, detections should be temporally ordered
            if current_local_time[:8] == system_pred.local_time[i][:8] and current_track_id == system_pred.track_id[i]:
                assert int(current_local_time) < int(system_pred.local_time[i])
                
                # a track spans two UTC days
                if current_UTC_date != system_pred.filename[i][4:12]:
                    print(station, system_pred.local_time[i], system_pred.track_id[i])

            # new track, the local date should be the same or bigger
            else:
                assert int(current_local_time[:8]) <= int(system_pred.local_time[i][:8])
                
            current_local_time = system_pred.local_time[i]
            current_track_id = system_pred.track_id[i]
            current_UTC_date = system_pred.filename[i][4:12]
```
